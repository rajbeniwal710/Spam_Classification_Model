Remove Punctuation

```
import string
string.punctuation
```

```
def remove_punct(text):
  non_punct_text = "".join([char for char in text if char not in string.punctuation])
  return non_punct_text

full_corpus['clean_text'] = full_corpus['text'].apply(lambda x: remove_punct(x))
```

```
full_corpus.head()
```

Tokenization

```
def tokenize(text):
  tokens = re.split('\W+', text)
  return tokens

full_corpus['text_token'] = full_corpus['clean_text'].apply(lambda x: tokenize(x.lower()))
```

```
full_corpus.head()
```

Remove stopwords

```
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
```

```
def remove_stop(text):
  no_stop = [word for word in text if word not in stopwords]
  return no_stop

full_corpus['text_no_stop'] = full_corpus['text_token'].apply(lambda x: remove_stop(x))
full_corpus.head()
```

Data Cleaning
Stemming

```
ps = nltk.PorterStemmer()
```

```
def stemming_text(text):
  text_stem = [ps.stem(word) for word in text]
  return text_stem

full_corpus['text_stemmed'] = full_corpus['text_no_stop'].apply(lambda x: stemming_text(x))
full_corpus.head()
```

text cleaning using lemmatization

```
nltk.download('wordnet')
wn = nltk.WordNetLemmatizer()
```

```
def lematize_text(text):
  text_stem = [wn.lemmatize(word) for word in text]
  return text_stem

full_corpus['text_lemmatized'] = full_corpus['text_no_stop'].apply(lambda x: lematize_text(x))
full_corpus.head()
```

Vextorization
its types are
1. count vectorization
2. n-gram vectorization
3. inverse document frequency weighting

```
data = full_corpus[['label', 'text','clean_text']]
```

```
def clean_text(text):
  text1 = "".join([char.lower() for char in text if char not in string.punctuation])
  tokens = re.split('\W+', text1)
  text_stem = [wn.lemmatize(word) for word in tokens if word not in stopwords]
  return text_stem
```

```
#count vectorization
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(analyzer = clean_text)
x_counts = count_vect.fit_transform(data['text'])
print(x_counts.shape)
print(count_vect.get_feature_names())
```

```
#n-gram vectorization
ngram = CountVectorizer(ngram_range=(2,2))
x_counts = ngram.fit_transform(data['clean_text'])
print(x_counts.shape)
print(ngram.get_feature_names())
```

```
#tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(analyzer = clean_text)
x_count = tfidf.fit_transform(data['text'])
print(x_count.shape)
print(tfidf.get_feature_names())
```
